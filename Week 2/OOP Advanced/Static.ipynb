## **Static Keyword**

```python
class Animal {
 
    // method 
    void display() {
        System.out.println("I am an animal.");
    }
    
    // main method
     public static void main(String[] args) {
        
         // object of Animal class
         Animal obj = new Animal();
         
         // access the method of Animal class
         obj.display();
    }
}

// I am an animal.
```

Đôi khi có những trường hợp ta cần **truy cập các trường và phương thức mà không muốn tạo đối tượng**. Khi ấy, chúng ta có thể sử dụng từ khóa **static**.

### **Example**

```python
class Animal {
 
    // method 
    static void display() {
        System.out.println("I am an animal.");
    }
    
    // main method
    public static void main(String[] args) {
         
         // access the method of Animal class
         display();
    }
}
 
// Output: I am an animal.
```

Chúng ta có thể truy cập trực tiếp phương thức **display()** mà không cần sử dụng đối tượng của lớp. Đó là bởi vì ta đã xác định phương thức là **static**.

```python
class Animal {
    
    static String type = "Mammal";
    
    // main method
     public static void main(String[] args) {
         
         // access the method of Animal class
         System.out.println("Animal Type: " + type);
    }
}
 
// Output: Animal Type: Mammal
```

### **Why use static?**

Khi khai báo một biến **static**, **tất cả các đối tượng của lớp đều có biến static đó**. 

**Các biến và phương thức static thuộc về lớp (chứ không phải đối tượng)**. Và chúng ta không cần tạo các đối tượng của lớp để truy cập các biến và phương thức static.

```python
class Company {
    
    static String name;
}
 
class Main {
    
    public static void main(String[] args) {
        
        Company.name = "Programiz";
    }
}
```

**Tất cả các đối tượng của lớp Company đều có chung một trường static name.**

Tuy nhiên, khi khai báo một biến không static, tất cả các đối tượng sẽ có các bản sao riêng biệt của biến.

```python
class Company {
    
    String name;
}
 
class Main {
    
    public static void main(String[] args) {
        
        Company object1 = new Company();
        Company object2 = new Company();
    }
}
```

Cả hai đối tượng c1 và c2 sẽ có các bản sao riêng biệt của biến name và chúng hoàn toàn khác nhau.

### **Accessing static fields/methods from other classes**

Giả sử chúng ta có một phương thức **static display()** bên trong lớp Animal và chúng ta phải truy cập nó từ lớp Main. Hãy cùng xem cách thực hiện:

```python
class Animal {
 
    // static method
    static void display() {
        System.out.println("I am an animal.");
    }
 
}
 
class Main {
 
    // main method
    public static void main(String[] args) {
 
        // access the method of Animal class
        Animal.display();
    }
}
 
// Output
// I am an animal.
```
