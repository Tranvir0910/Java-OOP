## **Class**

### **Create Object**

```python
// create a class
class Student {

    public static void main(String[] args) {
    
        // create objects of Student class
        Student student1 = new Student();
        Student student2 = new Student();

    }
}
```

### **Access field using object**

```python
class Student {
 
    // add fields
    String name = "Jackie";
    int score = 40;
    
    public static void main(String[] args) {
        
        // create object of Student
        Student student1 = new Student();
        
        // access fields using object
        System.out.println("Name: " + student1.name);
        System.out.println("Score: " + student1.score);
    }
}
```

### **Using classes and objects**

#### **Gán giá trị cho trường sử dụng đối tượng**

```python
class Student {
   
    // add fields
    String name = "Jackie";
    int score = 40;
}
```

Theo hướng tiếp cận OOP, đây không phải là cách sử dụng các trường khuyên dùng.


Thay vào đó, ta chỉ nên khai báo trường bên trong lớp và gán giá trị sử dụng đối tượng.

```python
class Student {
    
    // define fields
    String name;
    int score;
    
    public static void main(String[] args) {
        
        // create object of Student
        Student student1 = new Student();
        
        // assign values to fields
        student1.name = "Maria";
        student1.score = 56;
        
        // access fields
        System.out.println("Name: " + student1.name);
        System.out.println("Score: " + student1.score);
    }
}
```

Trong ví dụ trên, ta đã tạo một đối tượng của **lớp Student**.

```python
Student student1 = new Student();
```

Sau đó, ta sử dụng đối tượng **student1** với toán tử dấu chấm **.** để gán giá trị cho các trường: **name** và **score**.

```python
student1.name = "Maria";
student1.score = 56;
```

#### **Sửa đổi giá trị trường**

```python
class Student {
    String name;
    
    public static void main(String[] args) {
        
        Student bestStudent = new Student();
        
        // assign value to name
        bestStudent.name = "Jack";
        System.out.println("Best Student: " + bestStudent.name);
        
        // change the value of name
        bestStudent.name = "Rosa";
        System.out.println("Best Student: " + bestStudent.name);
    }
}
```

```python
Output:
Best Student: Jack
Best Student: Rosa
```

Giá trị ban đầu của **name** là **"Jack"** và sau đó được thay đổi thành **"Rosa"**.

#### **Tạo nhiều đối tượng**

```python
class Student {

   // define fields
   String name;
   int score;

   public static void main(String[] args) {

       // create object of Student class
       Student student1 = new Student();

       // field values for student1
       student1.name = "Maria";
       student1.score = 56;
       System.out.println("Name: " + student1.name);
       System.out.println("Score: " + student1.score);

       // create another object of Student class
       Student student2 = new Student();
      
       // field values for student2
       student2.name = "Johnny";
       student2.score = 32;
       System.out.println("Name: " + student2.name);
       System.out.println("Score: " + student2.score);
   }
}
```

```python
Output:

Name: Maria
Score: 56
Name: Johnny
Score: 32
```

### **Example**

```python
public class Point {

    // Declaration of two fields: x and y
    private int x;      
    private int y;

    // Declaration of a constructor     
    public Point(int startx, int starty){          
        X = startx;
        y = starty;
    }

    // Declaration of three methods
    public int getX() { return X; }
    public int getY() { return y; }
    public double distance (Point otherPt) {
        int dx = x - otherPt.getX();
        int dy = y - otherPt.getY();
        return Math.sqrt (dx * dx + dy * dy);
    }

    // main
    public static void main (String[] args) {
        Point p1 = new Point (3,4);
        Point p2 = new Point (6,8);
        System.out.println(p1.distance(p2));
    }
}
```

```python
Bắt đầu
|
|--- Tạo lớp Point
|   |
|   |--- Khai báo hai trường: x và y (kiểu private int)
|   |
|   |--- Khai báo constructor (public Point(int startx, int starty))
|   |   |
|   |   |--- Gán startx vào trường x
|   |   |--- Gán starty vào trường y
|   |
|   |--- Khai báo ba phương thức (public int, public int, public double)
|   |   |
|   |   |--- Phương thức getX: Trả về giá trị của trường x
|   |   |--- Phương thức getY: Trả về giá trị của trường y
|   |   |--- Phương thức distance: Nhận tham số otherPt là một đối tượng Point
|   |   |   |
|   |   |   |--- Tạo biến dx để tính khoảng cách theo trục x (x - otherPt.getX())
|   |   |   |--- Tạo biến dy để tính khoảng cách theo trục y (y - otherPt.getY())
|   |   |   |--- Trả về căn bậc hai của tổng bình phương của dx và dy
|
|--- Hàm main (public static void)
    |
    |--- Khởi tạo đối tượng p1 thuộc lớp Point với startx = 3 và starty = 4
    |--- Khởi tạo đối tượng p2 thuộc lớp Point với startx = 6 và starty = 8
    |--- In ra màn hình kết quả của phương thức distance của p1 với tham số otherPt là p2
Kết thúc
```

```python
Output:
5.0
```

- **Private** có nghĩa là chỉ có mã lệnh bên trong lớp này mới có thể trực tiếp thao tác với các trường này.

- Một **constructor** xác định cách tạo đối tượng của lớp này.

    + Nó là mã lệnh được thực thi khi một đối tượng được tạo ra để khởi tạo đối tượng đó. 

    + Lưu ý rằng **constructor** trông giống như một hàm **nhưng không có kiểu trả về và cùng tên với tên lớp**.

    + Trước constructor, chúng ta có từ **"public"**, có nghĩa là **bất kỳ mã lệnh nào cũng có thể sử dụng constructor này để tạo một điểm**.

- Sau **constructor**, chúng ta có **ba phương thức, getX, getY và distance**. Phương thức là các hàm nằm trong lớp. Trong Java, mọi thứ đều nằm trong một lớp, vì vậy kỹ thuật là **tất cả các hàm trong Java đều là phương thức**.

- Cuối cùng, chúng ta có khai báo của một **phương thức tĩnh (static)**. Các phương thức tĩnh hoạt động khác với phương thức thông thường. 

    + Chúng không tác động lên bất kỳ thể hiện cụ thể nào của một lớp, mà chỉ thuộc về lớp nói chung.

    + Phương thức này được gọi là "main", đây là một phương thức đặc biệt. Nếu bạn chạy chương trình của mình bên ngoài Blue Jay,

Ta có thể tạo một lớp chứa các trường và phương thức liên quan đến nhau và một lớp khác chứa **phương thức main()** để **truy cập** và **sửa đổi** các trường và phương thức của lớp trước đó.

```python
//Student Class
class Student {
 
    String name;
    int score;
 
    void checkPassFail(String name, int score) {
 
        if (score >= 35) {
            System.out.println(name + " passed the exam.");
        } else {
            System.out.println(name + " failed the exam.");
        }
    }
}
 
//Main Class
class Main {
    public static void main(String[] args) {
 
        // create first object
        Student student1 = new Student();
        student1.name = "Jack";
        student1.score = 78;
 
        // call checkPassFail with name and score of student1
        student1.checkPassFail(student1.name, student1.score);
 
        // create second object
        Student student2 = new Student();
        student2.name = "Rosa";
        student2.score = 32;
 
        // call checkPassFail with name and score of student2
        student2.checkPassFail(student2.name, student2.score);
    }
}
```

```python
Output

Jack passed the exam.
Rosa failed the exam.
```

**Lớp Student**

- Bao gồm hai trường **name** và **score** và một **phương thức checkPassFail()**.

- Phương thức nhận 2 tham số và in ra sinh viên đỗ hay trượt.

**Lớp Main**

- Chúng ta **gán** các giá trị **name** và **score** cho hai đối tượng **student1** và **student2**.

- Sau đó, ta gọi phương thức **checkPassFail()** với giá trị **name** và **score** của cả hai đối tượng.

 Note
----

- Chúng ta thêm **phương thức main()** bên trong **lớp Main**, vì vậy việc thực thi chương trình bắt đầu từ đó.

- Nếu ta chưa truy cập các trường và phương thức của **lớp Student** từ **main()** thì **lớp Student** sẽ trở nên vô dụng.
