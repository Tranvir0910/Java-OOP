## **Getter and Setter**

- **Phương thức Setter** - cho phép thiết lập giá trị của trường

- **Phương thức Getter** - cho phép lấy giá trị của trường

```python
class Student{
    private String name;
}
 
class Main {
 
    public static void main(String[] args) {
 
        // create objects of Student class
        Student student1 = new Student();
   
        // Error Code
        student1.name = "Felix";
        System.out.println(student1.name);
    }
}
```

Code này sẽ gây ra lỗi vì ta đang cố truy cập trực tiếp vào trường private từ lớp khác.

Bây giờ, sử dụng **getter** và **setter** để truy cập trường name này.

```python
class Student {
    private String name;
 
    // setter method
    public void setName(String name) {
        this.name = name;
    }
 
    // getter method
    public String getName() {
        return this.name;
    }
}
 
class Main {
 
    public static void main(String[] args) {
 
        // create objects of Student class
        Student student1 = new Student();
 
        // assign value using the setter method
        student1.setName("Alex");
 
        // access the value using getter
        String name = student1.getName();
        System.out.println("Name: " + name);
    }
}
```
