## **Public**

### **Chỉ định truy cập (access modifier)**

Chỉ định truy cập được sử dụng để **thiết lập mức truy cập cho các trường/lớp**


Nếu không muốn một số trường/phương thức được truy cập từ các lớp khác, ta thể đánh dấu trường là **private** bằng cách sử dụng **chỉ định truy cập private**.

```python
class Student {

    private int age;

    private void displayName() {
        ...   
    }
}
```

Trong bài học này, chúng ta sẽ tìm hiểu về hai loại chỉ định truy cập chính trong Java.

- **public** - cho phép truy cập từ các lớp khác.

- **private** - ngăn truy cập từ các lớp khác.

### **Public modifier**

```python
class Student{
 
    // public field 
    public String name;
}
 
class Main {
 
    public static void main(String[] args) {
 
        // create objects of Student class
        Student student1 = new Student();
   
        // access the public field from the Main class
        student1.name = "Felix";
        System.out.println(student1.name);
    }
}
 
// Output: Felix
```

Trong ví dụ trên, ta đã tạo chỉ định truy cập **public** với biến **name**. 

Đó là lý do tại sao ta có thể gán một giá trị mới và truy cập giá trị của nó từ lớp **Main**.

#### **Phương thức public**

```python
class Student{
 
    // public field 
    public void displayInfo() {
        System.out.println("I am a Student");
    }
}
 
class Main {
 
    public static void main(String[] args) {
 
        // create objects of Student class
        Student student1 = new Student();
   
        // access the public method from the Main class
        student1.displayInfo();
    }
}
 
// Output: I am a Student
```

Như bạn thấy, chúng ta có thể truy cập phương thức public của **lớp Student** từ **lớp Main**.

#### **Phương thức main()**

Hãy nhớ rằng phương thức main() cũng có phạm vi truy cập là public.


```python
public static void main(String[] args)
```

Điều này là do trình biên dịch Java gọi phương thức main() để thực thi chương trình. 

Vì vậy, để cho phép trình biên dịch gọi phương thức, nó phải có phạm vi **public**.

### **Private modifier**

```python
class Student {
 
    private String name;
}
 
class Main {
 
    public static void main(String[] args) {
 
        // create objects of Student class
        Student student1 = new Student();
 
        // trying to access the private field
        student1.name = "Felix";
        System.out.println(student1.name);
    }
}
```

Khi chạy code trên, chúng ta sẽ gặp lỗi:

```python
error: name has private access in Student
student1.name = "Felix";
                   ^
```

Có thể thấy chương trình gặp lỗi vì ta đang cố gắng truy cập trường **private name** từ **lớp Main**.

### **Phương thức private**

```python
class Student {
    
    private void displayName() {
        System.out.println("I am Private Method");
    }
}
class Main {
    public static void main(String[] args) {
 
        Student student1 = new Student();
 
        // try to access the private method
        student1.displayName();
    }
}
```

Khi chạy code trên, chúng ta sẽ gặp lỗi tương tự:

```python
error: displayName() has private access in Student
student1.displayName();
                ^
```
